```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import cv2
from dask.array import linalg
```

```python
data = fits.open('../data/rosat/Rosat-skv14717560352171.fits')
data.info()
```

```python
img = data[0].data[0:200, 50:250]
```

```python
blur = cv2.GaussianBlur(img,(21,21),0)
```

```python
x0, y0 = 276.14, -24.87
```

```python
x, y = x0 + np.linspace(-0.5, 0.5, 300) * 3.75, y0 + np.linspace(-0.5, 0.5, 300) * 3.75
x1, y1 = x[0:200], y[50:250]
plt.figure(figsize=(8, 8))
plt.pcolormesh(x1, y1, blur)
plt.scatter(274.84, -25.407, color='cyan', marker='*')
plt.gca().invert_xaxis()
plt.gca().set_aspect('equal')
plt.show()
```
